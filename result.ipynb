```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv('cleaned.csv')
```

```python
data.head(10)
```

```python
data['time'] = pd.to_datetime(data['time'])
```

```python
data.info()
```

```python
df0 = data[data['abnormality'] == 0]
df1 = data[data['abnormality'] == 1]
```

```python
df0.info()
df1.info()
```

```python
df0.set_index("time", inplace = True)
df1.set_index("time", inplace = True)
```

```python
cols = data.drop(['time', 'abnormality'], axis=1).columns
```

```python
for col in cols:
    plott = plt.figure(figsize=(30,10))
    plott = plt.scatter(df0.index, df0[col], c = "blue", marker='.')
    plott = plt.scatter(df1.index, df1[col], c = "red", marker='x')
    plott = plt.title(col)
    plott = plt.show()
```
